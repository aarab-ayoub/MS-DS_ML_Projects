# Perceptron avec Perceptron de sklearn et Dataset mails
Objectif c'est de classifier les emails (<b>spam</b> ou <b>non spam</b>) avec <b>Perceptron</b> de <b>sklearn</b>

```python
import pandas as pd #Importer les Dataset
import numpy as np #Manipuler le Dataset comme matrice
import matplotlib.pyplot as plt #Visualiser Les données
from sklearn.linear_model import Perceptron #Contient l'algorithme de Perceptron
```

```python
#charger le dataset
df = pd.read_csv('Dataset/mails.csv',delimiter=";")
df.head()
```

```python
#X = df.iloc[:,:4].values

X1 = df.iloc[:,0].values #colonne des fautes d'orthographe 
X2 = df.iloc[:,1].values #colonne des nbr des mots-clés déclencheurs
Y = df.iloc[:,-1].values #colonne spam ou non spam

X1 = X1.reshape(len(X1),1) #redimenssionner X1 vers un vecteur colonne
X2 = X2.reshape(len(X2),1)
X = np.hstack((X1, X2)) #Concatenation des features

Y = Y.reshape(len(Y),1)

X11 = X1[Y == 1]#Contient que les X1 dont y == 1
X12 = X1[Y == 0]#Contient que les X1 dont y == 0
X21 = X2[Y == 1]#Contient que les X2 dont y == 1
X22 = X2[Y == 0]#Contient que les X2 dont y == 0

#y = y.reshape(len(y),1)
#plt.title('Spam ou Non Spam')

plt.xlabel("x1 : fautes d'orthographe")
plt.ylabel("x2 : nbr des mots-clés déclencheurs")

plt.plot(X11, X21, 'xr')
plt.plot(X12, X22, 'ob')
x = np.linspace(1,7)
y = -x + 7
plt.text(5,2, '$x_1 + x_2 = 7$', color = 'm',fontsize = 20)
plt.text(3, 4, "$ ~~~~~~ \hat{y} = 1$", color = 'r',fontsize = 20)
plt.arrow(3,4,1,1,width=0.1,edgecolor='red', facecolor='red')
plt.text(1, 2.2, "$ ~~~~~~ \hat{y} = 0$", color = 'c', fontsize = 20)
plt.arrow(3,4,-1,-1,width=0.1,edgecolor='c', facecolor='c')
plt.plot(x, y, 'm')
```

```python
Y
```

### Matrice X des features
Construire la matrice des <b>features</b> + une colonne des <b>1</b>

$X = \begin{pmatrix} x_1^{(1)} x_2^{(1)} 1\\ x_1^{(2)} x_2^{(2)} 1\\...\\x_1^{(m)} x_2^{(m)} 1 \end{pmatrix}$

```python
X = np.hstack((X1, X2)) #Concatenation des features
```

```python
X = np.hstack((X, np.ones(X1.shape))) #Ajouter une colonne des 1
```

```python
X
```

### Création du modèle d'un perceptron
Pour ce faire : Perceptron()

```python
# Création du modele d'un perceptron
modele = Perceptron(max_iter=1000, eta0=0.01)
```

```python
#Entrainement
modele.fit(X, Y)
```

```python
#Coefficient de détermination pour tester la performance du modèle
modele.score(X, Y)
```

### Affichage de la frontière de décision
### Lorsque $F(X) = w_1 \times x_1 + w_2 \times x_2 + b = 0$
### càd $x_2 = - \frac{w_1}{w_2} \times x_1   - \frac{b}{w_2}$ 
avec $w_1 = \theta[0]$, $w_2 = \theta[1]$ et $b = \theta[2]$

```python
plt.xlabel("x1")
plt.ylabel("x2")
plt.plot(X1, X2, 'x')

plt.plot(X[:, 0], X[:, 1], 'x')

#w1, w2 et b ou bien theta0, theta1, theta2
theta0, theta1, theta2 = modele.coef_[0][0], modele.coef_[0][1], modele.intercept_[0]

plt.plot(X[:, 0], -theta0*X[:, 0]/theta1 - theta2/theta1, 
         label = 'Frontière de décision : $G(X) = w_1 \\times x_1 + w_2 \\times x_2 + b = 0$', color='m')
plt.legend()
#plot_x = np.array([min(X[:, 0]), max(X[:, 0])])
```

### Prédiction
### Exp1 :
email1 : fautes d'orthographe = 6 	nbr des mots-clés déclencheurs = 6

### Prédiction :
<b>Spam</b>

### Exp2 :
email2 : fautes d'orthographe = 1 	nbr des mots-clés déclencheurs = 1

### Prédiction :
<b>Non spam</b>

```python
print(modele.predict([[6, 6, 1]])) #proche de 1
```
